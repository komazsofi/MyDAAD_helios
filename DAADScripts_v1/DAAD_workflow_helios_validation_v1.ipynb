```python
import numpy as np
import matplotlib.pyplot as plt

v=np.array([10,20,30,41,51])
pdens_measured=np.array([38,20,12.5,10,8])
pdens_sim=np.array([40,19.96,13.68,9.86,7.31])

plt.plot(v,pdens_measured,linewidth=2,label='Riegl data sheet')
plt.plot(v,pdens_sim,linewidth=2,label='Simulated')
plt.scatter(v,pdens_measured)
plt.scatter(v,pdens_sim)
plt.xlabel('Velocity [m/s]')
plt.ylabel('Point density [pts/m2]')
plt.legend()
plt.title('ALS [PRR=240 kHz, AGL=380 m, FOV=45 degree]')
plt.show()

```
