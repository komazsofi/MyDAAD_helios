1.) Import 

Imported file structure is the following: [filename,platform,treemodellid,feature1..featuren]

```python
import pandas as pd
import numpy as np

workdir='/home/komazsofi/Munka/Heidelberg/DAAD_helios_sim/Data2/'
filename='Sim_obj_geom_fea.txt'
feature='mean_z'

sim_obj_geomfea=pd.read_csv(workdir+filename,sep=',',names=['filename','platform','treemodellid','shape','scanposition','mean_z','median_z','var_z','std_z','coeffvar_z','skew_z','kurto_z','return'])
```

2.) Filter by platform

```python
sim_obj_geomfea_als=sim_obj_geomfea[(sim_obj_geomfea['platform']=='ALS')]
sim_obj_geomfea_tls=sim_obj_geomfea[(sim_obj_geomfea['platform']=='TLS')]
sim_obj_geomfea_mls=sim_obj_geomfea[(sim_obj_geomfea['platform']=='MLS')]
sim_obj_geomfea_uls=sim_obj_geomfea[(sim_obj_geomfea['platform']=='ULS')]
```

3.) Explorative analysis

a.) Boxplot visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x=feature, y="shape", hue="return", data=sim_obj_geomfea,palette="RdBu_r")
plt.show()

sns.boxplot(x=feature, y="shape", hue="platform", data=sim_obj_geomfea)
plt.show()

sns.boxplot(x=feature, y="platform", data=sim_obj_geomfea)
plt.show()

sns.boxplot(x=feature, y="platform",hue='shape', data=sim_obj_geomfea,palette="RdBu_r")
plt.show()
```

b.) Jointplot (https://ocefpaf.github.io/python4oceanographers/blog/2014/06/16/seaborn/)

```python
sns.jointplot(sim_obj_geomfea_als[feature].values, sim_obj_geomfea_tls[feature].values, kind="reg")
plt.show()
```

c.) Histogram

```python
plt.hist([sim_obj_geomfea_als[feature].values,sim_obj_geomfea_tls[feature].values,sim_obj_geomfea_mls[feature].values,sim_obj_geomfea_uls[feature].values],10,histtype='bar',label=['als', 'tls', 'mls','uls'])
plt.legend()
plt.show()
```
